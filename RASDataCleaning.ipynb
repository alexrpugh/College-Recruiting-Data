## RAS Data Exploration and Cleaning

```python
import pandas as pd
pd.set_option('display.max_columns', None)

RASData = pd.read_csv("/Users/alexpugh/UConn 2024 Spring/Football/data/RASData.csv")
```

```python
RASData.replace({"NULL": pd.NA, "PLACEHOLDER": pd.NA, "XX": pd.NA}, inplace=True)
```

```python
# Drop columns with only NaN values
RASData.dropna(axis=1, how='all', inplace=True)
# Drop the first row
RASData = RASData.drop(0)
# Drop rows with at least 10 NaN values
RASData = RASData.dropna(thresh=RASData.shape[1] - 12 + 1)
```

```python
# Drop all players from 2024 NFL Combine due to absence of testing numbers
RASData = RASData[RASData['Year'] != '2024']
RASData
```
